```python
# Template classın class tanımlaması. Yalnızca data member ve bu 4 metod protipi
template <class T>
class TemplateClass
{
private:
    T dataMember;

public:
    TemplateClass();
    void setData(T value);
    T getData() const;
    void display() const;
};

# İnsert metodu
template <class T>
void TemplateClass<T>::setData(T value)
{
    dataMember = value;
}
```

```python
# Yeni array oluşturup boyutu 2 katına çıkarma fonkunu resize metoduna yazma
template <class T>
void TemplateClass<T>::resize()
{
    T* newArray = new T[size * 2];

    for (int i = 0; i < size; i++)
    {
        newArray[i] = dataMember[i];
    }

    delete[] dataMember;
    dataMember = newArray;
    size *= 2;
}
```

```python
# Dequeue in içeriği
public T Dequeue()
{
    if (size == 0)
        throw new InvalidOperationException();

    T value = array[head];   // 1. baştaki elemanı al

    array[head] = default;   // 2. temizle (GC için)

    head = (head + 1) % array.Length;  // 3. head pointer ilerlet (circular)

    size--;  // 4. eleman sayısını azalt

    return value;  // 5. değeri döndür
}

# Enqueue nin içeriği
public void Enqueue(T item)
{
    if (size == array.Length)
    {
        // kapasite doluysa büyüt
        int newCapacity = array.Length == 0 ? 4 : array.Length * 2;

        T[] newArray = new T[newCapacity];

        if (size > 0)
        {
            if (head < tail)
            {
                Array.Copy(array, head, newArray, 0, size);
            }
            else
            {
                Array.Copy(array, head, newArray, 0, array.Length - head);
                Array.Copy(array, 0, newArray, array.Length - head, tail);
            }
        }

        array = newArray;
        head = 0;
        tail = size;
    }

    array[tail] = item;            // 1. sona ekle
    tail = (tail + 1) % array.Length; // 2. circular ilerlet
    size++;                        // 3. eleman sayısını artır
}
```

```python
# Queue yapısı için basit bir class data member metod prototipi(kuyruk yapısı array değil ptr based olan)
template <class T>
class Queue
{
private:
    struct Node
    {
        T data;
        Node* next;
    };

    Node* front;
    Node* rear;
    int size;

public:
    Queue();
    void enqueue(T value);
    T dequeue();
    T peek() const;
    bool isEmpty() const;
};
```

```python
# Random bir arraydan maxı bulan fonksiyon ve karmaşıklığı
int findMax(int arr[], int n)
{
    int maxVal = arr[0];

    for (int i = 1; i < n; i++)
    {
        if (arr[i] > maxVal)
            maxVal = arr[i];
    }

    return maxVal;
}
# Time Complexity: O(n)
# Space Complexity: O(1)
```

```python
# Brute force, enumaration all case. Max sub problemi zaman karmaşıklığı O(n^3) olan algoritma
int maxSubArray(int arr[], int n)
{
    int maxSum = arr[0];

    for (int i = 0; i < n; i++)
    {
        for (int j = i; j < n; j++)
        {
            int sum = 0;

            for (int k = i; k <= j; k++)
            {
                sum += arr[k];
            }

            if (sum > maxSum)
                maxSum = sum;
        }
    }

    return maxSum;
}
```
