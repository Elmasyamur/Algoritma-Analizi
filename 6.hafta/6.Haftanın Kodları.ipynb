```python
# 6. hafta derste yazdığımız promptlar ve gördüğümüz örnek kodlar
# C# ta ctor complex olan bir örnek
using System;
using System.Collections.Generic;
using System.Linq;

namespace ConstructorComplexExample
{
    public class Employee
    {
        public string FullName { get; }
        public string Department { get; }
        public decimal BaseSalary { get; }
        public decimal TaxRate { get; }
        public List<string> Skills { get; }
        public string Email { get; }
        public DateTime CreatedAt { get; }

        // Complex ctor
        public Employee(
            string fullName,
            string department,
            decimal baseSalary,
            decimal taxRate,
            List<string> skills,
            string email = null)
        {
            // Validation
            if (string.IsNullOrWhiteSpace(fullName))
                throw new ArgumentException("Ad soyad boş olamaz.");

            if (baseSalary <= 0)
                throw new ArgumentException("Maaş 0'dan büyük olmalı.");

            if (taxRate < 0 || taxRate > 1)
                throw new ArgumentException("Vergi oranı 0 ile 1 arasında olmalı.");

            if (skills == null || !skills.Any())
                throw new ArgumentException("En az 1 skill girilmeli.");

            // Normalize data
            FullName = fullName.Trim();
            Department = department?.Trim().ToUpper();
            BaseSalary = baseSalary;
            TaxRate = taxRate;

            Skills = skills
                .Where(x => !string.IsNullOrWhiteSpace(x))
                .Select(x => x.Trim().ToLower())
                .Distinct()
                .ToList();

            // Auto generate email if null
            Email = string.IsNullOrWhiteSpace(email)
                ? GenerateEmail(FullName)
                : email.Trim().ToLower();

            CreatedAt = DateTime.Now;
        }

        private string GenerateEmail(string name)
        {
            return name.ToLower().Replace(" ", ".") + "@company.com";
        }

        public decimal NetSalary()
        {
            return BaseSalary - (BaseSalary * TaxRate);
        }

        public void Print()
        {
            Console.WriteLine($"Ad: {FullName}");
            Console.WriteLine($"Departman: {Department}");
            Console.WriteLine($"Brüt Maaş: {BaseSalary}");
            Console.WriteLine($"Net Maaş: {NetSalary()}");
            Console.WriteLine($"Email: {Email}");
            Console.WriteLine($"Skills: {string.Join(", ", Skills)}");
            Console.WriteLine($"Oluşturulma: {CreatedAt}");
        }
    }

    class Program
    {
        static void Main()
        {
            Employee emp = new Employee(
                fullName: " Ahmet Yılmaz ",
                department: "software",
                baseSalary: 50000,
                taxRate: 0.20m,
                skills: new List<string>
                {
                    "C#",
                    "SQL",
                    "C#",
                    " ASP.NET "
                }
            );

            emp.Print();
        }
    }
}
```

```python
# Java person classtan bir array örneği
class Person {
    String name;

    Person(String name) {
        this.name = name;
    }
}

public class Main {
    public static void main(String[] args) {
        Person[] people = {
            new Person("Ahmet"),
            new Person("Mehmet"),
            new Person("Ayşe")
        };

        System.out.println(people[0].name); // Ahmet
    }
}
```

```python
# Değer girmeyi sonra yaparsak(ctor da değil)
class Person {
    String name;
}

public class Main {
    public static void main(String[] args) {
        Person[] people = new Person[3];

        people[0] = new Person();
        people[0].name = "Ahmet";

        people[1] = new Person();
        people[1].name = "Mehmet";

        people[2] = new Person();
        people[2].name = "Ayşe";

        System.out.println(people[1].name); // Mehmet
    }
}

# 4.personu eklersek

people[3] = new Person();
people[3].name = "Veli";

# burada new Person[3] olduğu için ArrayIndexOutOfBoundsException hatası döner
```

```python
# Java da int array ama resize yapan küçük bir class yazma örneği
class IntList {
    private int[] data;
    private int size;

    public IntList() {
        data = new int[2]; // başlangıç kapasitesi
        size = 0;
    }

    public void insert(int value) {
        if (size == data.length) {
            // resize
            int[] newData = new int[data.length * 2];

            for (int i = 0; i < data.length; i++) {
                newData[i] = data[i];
            }

            data = newData;
        }

        data[size] = value;
        size++;
    }

    public void print() {
        for (int i = 0; i < size; i++) {
            System.out.println(data[i]);
        }
    }
}
```
