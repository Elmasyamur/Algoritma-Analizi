```python
"""
Aşağıdaki fonksiyon, rastgele bir dizi üretip içinde minimum ve maksimum
değerleri, onların indexlerini ve diziyi tararken yaptığı toplam karşılaştırma 
sayısını hesaplar ve hepsini tek seferde döndürür.
"""
import random

def dizi_min_max_step_sayac(n, alt_sinir=0, ust_sinir=100):
    
    # Rastgele dizi oluştur
    dizi = [random.randint(alt_sinir, ust_sinir) for _ in range(n)]

    # Başlangıç değerleri
    min_deger = dizi[0]
    max_deger = dizi[0]
    min_index = 0
    max_index = 0

    # Arama sayacı
    arama_sayisi = 0

    # Döngü
    for i in range(len(dizi)):
        
        arama_sayisi += 1
        if dizi[i] < min_deger:
            min_deger = dizi[i]
            min_index = i

        arama_sayisi += 1
        if dizi[i] > max_deger:
            max_deger = dizi[i]
            max_index = i

    return dizi, min_deger, min_index, max_deger, max_index, arama_sayisi


# Örnek kullanım
dizi, min_val, min_i, max_val, max_i, arama = dizi_min_max_step_sayac(10)

print("Dizi:", dizi)
print("Minimum:", min_val, "Index:", min_i)
print("Maksimum:", max_val, "Index:", max_i)
print("Toplam karşılaştırma sayısı:", arama)
```

```python
# yukarıdaki fonksiyonun yalnızca min olduğu hali
import random

def dizi_min_step_sayac(n, alt_sinir=0, ust_sinir=100):

    # Rastgele dizi oluştur
    dizi = [random.randint(alt_sinir, ust_sinir) for _ in range(n)]

    # Başlangıç değerleri
    min_deger = dizi[0]
    min_index = 0

    # Karşılaştırma sayacı
    arama_sayisi = 0

    # Döngü
    for i in range(len(dizi)):
        arama_sayisi += 1

        if dizi[i] < min_deger:
            min_deger = dizi[i]
            min_index = i

    return dizi, min_deger, min_index, arama_sayisi


# Örnek kullanım
dizi, min_val, min_i, arama = dizi_min_step_sayac(10)

print("Dizi:", dizi)
print("Minimum değer:", min_val)
print("Minimum index:", min_i)
print("Toplam karşılaştırma:", arama)
```

```python
# 2. min değerini bulan fonksiyon
import random

def ikinci_min_bul(n, alt_sinir=0, ust_sinir=100):

    # Rastgele dizi
    dizi = [random.randint(alt_sinir, ust_sinir) for _ in range(n)]

    # Başlangıç değerleri
    min1 = float('inf')
    min2 = float('inf')

    min1_index = -1
    min2_index = -1

    arama_sayisi = 0

    for i in range(len(dizi)):

        arama_sayisi += 1

        if dizi[i] < min1:
            min2 = min1
            min2_index = min1_index

            min1 = dizi[i]
            min1_index = i

        elif dizi[i] < min2 and dizi[i] != min1:
            min2 = dizi[i]
            min2_index = i

    return dizi, min1, min1_index, min2, min2_index, arama_sayisi


# Örnek kullanım
dizi, min1, i1, min2, i2, arama = ikinci_min_bul(10)

print("Dizi:", dizi)
print("1. Minimum:", min1, "Index:", i1)
print("2. Minimum:", min2, "Index:", i2)
#print("Toplam karşılaştırma:", arama)
```

```python
# 5. minimum değeri bulan fonksiyon
import random

def besinci_min_bul(n, alt_sinir=0, ust_sinir=100):

    # Rastgele dizi oluştur
    dizi = [random.randint(alt_sinir, ust_sinir) for _ in range(n)]

    # İlk 5 minimum için başlangıç değerleri
    mins = [float('inf')] * 5
    indexler = [-1] * 5

    arama_sayisi = 0

    for i in range(len(dizi)):

        arama_sayisi += 1
        sayi = dizi[i]

        for j in range(5):
            if sayi < mins[j]:
                # değerleri sağa kaydır
                for k in range(4, j, -1):
                    mins[k] = mins[k-1]
                    indexler[k] = indexler[k-1]

                mins[j] = sayi
                indexler[j] = i
                break

    return dizi, mins, indexler, mins[4], indexler[4], arama_sayisi


# Örnek kullanım
dizi, mins, idx, besinci_min, besinci_index, arama = besinci_min_bul(15)

print("Dizi:", dizi)
print("İlk 5 minimum:", mins)
print("Indexleri:", idx)
print("5. minimum:", besinci_min, "Index:", besinci_index)
print("Toplam arama:", arama)
```

```python
# Dizinin ilk 25 elemanını alıp bu elemanların insertion sort ile sıralanması ve 
# sonraki her eleman için eğer eleman 25. elemandan küçükse 
# 25'lik listede doğru yere insertion sort mantığı ile yerleştirip
# listenin son elemanını 25. minimum eleman olarak bulma yöntemi 

import random

def yirmibesinci_min(n, alt=0, ust=1000):

    # dizi üret
    dizi = [random.randint(alt, ust) for _ in range(n)]

    k = 25
    arama = 0

    # ilk 25 eleman
    A = dizi[:k]

    # insertion sort
    for i in range(1, k):
        key = A[i]
        j = i - 1

        while j >= 0 and A[j] > key:
            A[j+1] = A[j]
            j -= 1

        A[j+1] = key

    # geri kalan elemanlar
    for i in range(k, n):

        arama += 1

        if dizi[i] < A[k-1]:

            key = dizi[i]
            j = k - 2

            while j >= 0 and A[j] > key:
                A[j+1] = A[j]
                j -= 1

            A[j+1] = key

    return dizi, A, A[k-1], arama


# test
dizi, ilk25, min25, arama = yirmibesinci_min(100)

print("Dizi:", dizi)
print("En küçük 25 eleman:", ilk25)
print("25. minimum:", min25)
print("Karşılaştırma sayısı:", arama)
```

```python
# üstteki yöntemde insertion sort yerine bubble kullanarak yapalım

import random

def yirmibesinci_bubble(n, alt=0, ust=1000):
    # Rastgele dizi üret
    dizi = [random.randint(alt, ust) for _ in range(n)]
    
    arama_sayisi = 0
    # Bubble Sort
    for i in range(n):
        for j in range(0, n-i-1):
            arama_sayisi += 1
            if dizi[j] > dizi[j+1]:
                dizi[j], dizi[j+1] = dizi[j+1], dizi[j]
    
    # 25. minimum (index 24)
    m25 = dizi[24] if n >= 25 else None
    
    return dizi, m25, arama_sayisi

# Örnek kullanım
dizi, m25, arama = yirmibesinci_bubble(100)

print("Sıralanmış dizi:", dizi)
print("25. minimum:", m25)
print("Toplam karşılaştırma sayısı:", arama)
```

```python
# yukarıdaki yöntemde bu sefer bubble sort yerine selection sort kullanarak yapalım 
import random

def yirmibesinci_selection(n, alt=0, ust=1000):
    # Rastgele dizi oluştur
    dizi = [random.randint(alt, ust) for _ in range(n)]
    
    arama_sayisi = 0
    # Selection Sort
    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            arama_sayisi += 1
            if dizi[j] < dizi[min_idx]:
                min_idx = j
        # Takas
        dizi[i], dizi[min_idx] = dizi[min_idx], dizi[i]
    
    # 25. minimum
    m25 = dizi[24] if n >= 25 else None
    
    return dizi, m25, arama_sayisi


# Örnek kullanım
dizi, m25, arama = yirmibesinci_selection(100)

print("Sıralanmış dizi:", dizi)
print("25. minimum:", m25)
print("Toplam karşılaştırma sayısı:", arama)
```

```python
# bir diziyi selection, bubble ve insertion sort ile sıralayıp toplam karşılaştırma ve swap sayılarını yazdıran fonksiyon
# 3 sıralama algoritmasını karşılaştırmak adına bu fonksiyonu yazdık

import random

def rapor_sort_with_insertion(n, alt=0, ust=100):
    # Rastgele dizi üret
    orijinal_dizi = [random.randint(alt, ust) for _ in range(n)]

    # ----- Bubble Sort -----
    dizi_bubble = orijinal_dizi.copy()
    bubble_comparisons = 0
    bubble_swaps = 0

    for i in range(n):
        for j in range(0, n-i-1):
            bubble_comparisons += 1
            if dizi_bubble[j] > dizi_bubble[j+1]:
                dizi_bubble[j], dizi_bubble[j+1] = dizi_bubble[j+1], dizi_bubble[j]
                bubble_swaps += 1

    # ----- Selection Sort -----
    dizi_selection = orijinal_dizi.copy()
    selection_comparisons = 0
    selection_swaps = 0

    for i in range(n-1):
        min_idx = i
        for j in range(i+1, n):
            selection_comparisons += 1
            if dizi_selection[j] < dizi_selection[min_idx]:
                min_idx = j
        if min_idx != i:
            dizi_selection[i], dizi_selection[min_idx] = dizi_selection[min_idx], dizi_selection[i]
            selection_swaps += 1

    # ----- Insertion Sort -----
    dizi_insertion = orijinal_dizi.copy()
    insertion_comparisons = 0
    insertion_swaps = 0  # burada swap sayısı: her shift bir swap olarak sayılabilir

    for i in range(1, n):
        key = dizi_insertion[i]
        j = i - 1
        while j >= 0 and dizi_insertion[j] > key:
            insertion_comparisons += 1
            dizi_insertion[j+1] = dizi_insertion[j]  # shift
            insertion_swaps += 1
            j -= 1
        # son karşılaştırma kontrolü
        if j >= 0:
            insertion_comparisons += 1
        dizi_insertion[j+1] = key

    return {
        "orijinal_dizi": orijinal_dizi,
        "bubble": {
            "sıralı_dizi": dizi_bubble,
            "karşılaştırma": bubble_comparisons,
            "swap": bubble_swaps
        },
        "selection": {
            "sıralı_dizi": dizi_selection,
            "karşılaştırma": selection_comparisons,
            "swap": selection_swaps
        },
        "insertion": {
            "sıralı_dizi": dizi_insertion,
            "karşılaştırma": insertion_comparisons,
            "swap": insertion_swaps
        }
    }

# Örnek kullanım
rapor = rapor_sort_with_insertion(10)

print("Orijinal dizi:", rapor["orijinal_dizi"])
print("\n--- Bubble Sort ---")
print("Sıralı:", rapor["bubble"]["sıralı_dizi"])
print("Karşılaştırma:", rapor["bubble"]["karşılaştırma"])
print("Swap:", rapor["bubble"]["swap"])
print("\n--- Selection Sort ---")
print("Sıralı:", rapor["selection"]["sıralı_dizi"])
print("Karşılaştırma:", rapor["selection"]["karşılaştırma"])
print("Swap:", rapor["selection"]["swap"])
print("\n--- Insertion Sort ---")
print("Sıralı:", rapor["insertion"]["sıralı_dizi"])
print("Karşılaştırma:", rapor["insertion"]["karşılaştırma"])
print("Swap (shift sayısı):", rapor["insertion"]["swap"])
```

```python
# Her çalıştırmada farklı rastgele dizi ve hedef ile Linear ve Binary Search karşılaştırması
# için rastgele bir dizi üzerinde rastgele bir sayı arayalım

import random

def rastgele_arama(n, alt=0, ust=1000):
    # Rastgele dizi oluştur
    dizi = [random.randint(alt, ust) for _ in range(n)]
    
    # Diziyi sıralayalım (Binary Search için)
    dizi.sort()
    
    # Rastgele hedef seçelim
    target = random.randint(alt, ust)
    
    # ----- Linear Search -----
    linear_comparisons = 0
    linear_index = -1
    for i, val in enumerate(dizi):
        linear_comparisons += 1
        if val == target:
            linear_index = i
            break
    
    # ----- Binary Search -----
    binary_comparisons = 0
    low, high = 0, n - 1
    binary_index = -1
    while low <= high:
        mid = (low + high) // 2
        binary_comparisons += 1
        if dizi[mid] == target:
            binary_index = mid
            break
        elif dizi[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    
    # Sonuçları döndür
    return {
        "dizi": dizi,
        "target": target,
        "linear": {
            "index": linear_index,
            "karşılaştırma": linear_comparisons
        },
        "binary": {
            "index": binary_index,
            "karşılaştırma": binary_comparisons
        }
    }

# Örnek kullanım
rapor = rastgele_arama(20)

print("Sıralı dizi:", rapor["dizi"])
print("Aranan sayı:", rapor["target"])
print("\n--- Linear Search ---")
print("Index:", rapor["linear"]["index"])
print("Karşılaştırma sayısı:", rapor["linear"]["karşılaştırma"])
print("\n--- Binary Search ---")
print("Index:", rapor["binary"]["index"])
print("Karşılaştırma sayısı:", rapor["binary"]["karşılaştırma"])
```

```python
# random dizi üzerinde linear search uygulayalım

import random

def rastgele_linear_search(n, alt=0, ust=1000):
    # Rastgele dizi oluştur
    dizi = [random.randint(alt, ust) for _ in range(n)]
    
    # Rastgele hedef seçelim
    target = random.randint(alt, ust)
    
    # Linear Search
    linear_comparisons = 0
    linear_index = -1
    
    for i, val in enumerate(dizi):
        linear_comparisons += 1
        if val == target:
            linear_index = i
            break
    
    # Sonuçları döndür
    return {
        "dizi": dizi,
        "target": target,
        "index": linear_index,
        "karşılaştırma": linear_comparisons
    }

# Örnek kullanım
rapor = rastgele_linear_search(20)

print("Dizi:", rapor["dizi"])
print("Aranan sayı:", rapor["target"])
print("Bulunan index:", rapor["index"])
print("Toplam karşılaştırma sayısı:", rapor["karşılaştırma"])
```

```python
# rastgele dizi + Linear Search fonksiyonunu 100 kez çalıştırıp ortalama karşılaştırma
# sayısını hesaplama
import random

def rastgele_linear_search(n, alt=0, ust=1000):
    dizi = [random.randint(alt, ust) for _ in range(n)]
    target = random.randint(alt, ust)
    
    linear_comparisons = 0
    linear_index = -1
    
    for i, val in enumerate(dizi):
        linear_comparisons += 1
        if val == target:
            linear_index = i
            break
    
    return linear_comparisons

# ----- Ortalama Karşılaştırma Hesaplama -----
n = 20  # dizi boyutu
tekrar = 100
toplam_karşılaştırma = 0

for _ in range(tekrar):
    toplam_karşılaştırma += rastgele_linear_search(n)

ortalama_karşılaştırma = toplam_karşılaştırma / tekrar

print(f"{tekrar} tekrar için ortalama karşılaştırma sayısı:", ortalama_karşılaştırma)
```
