```python
import array

class IntArray:
    def __init__(self, size):
        # Belirtilen boyutta sıfırlarla dolu integer array
        self.arr = array.array('i', [0]*size)
        self.size = size
        self.count = 0  # mevcut dolu eleman sayısı

    def push_item(self, value):
        """Array'in sonuna eleman ekler"""
        if self.count < self.size:
            self.arr[self.count] = value
            self.count += 1
        else:
            print("Array dolu! Eleman eklenemedi.")

    def pop(self):
        """Array'in son elemanını çıkarır ve döndürür"""
        if self.count > 0:
            self.count -= 1
            value = self.arr[self.count]
            self.arr[self.count] = 0  # opsiyonel: boş olan yeri sıfırla
            return value
        else:
            print("Array boş! Eleman yok.")
            return None

    def __str__(self):
        # Sadece dolu elemanları göstermek için
        return str(self.arr[:self.count])
    

# Kullanım örneği
arr_obj = IntArray(5)
arr_obj.push_item(10)
arr_obj.push_item(20)
arr_obj.push_item(30)
print(arr_obj)  # [10, 20, 30]

popped = arr_obj.pop()
print("Pop edilen:", popped)  # 30
print(arr_obj)  # [10, 20]

# Yazdığımız yapı array-based bir yapıdır, pointer-based değil.
```

```python
arr_obj.push_item(10)  # integer
arr_obj.push_item(3.14)  # float eklemeye çalışırsak 
# TypeError: integer argument expected, got float fırlatır
# typed array’lerde eleman tipi sabittir; son elemanı başka tipe değiştirmek mümkün değil.
# bunu çözmek için:
arr = [1, 2, 3]
arr[-1] = "merhaba"  # son eleman tipi değişti
print(arr.pop())      # 'merhaba' çıkar
# merhabayı silmenin maaliyeti pop ile O(1) , liste ile O(n)
```

```python
# Python’da bir liste oluşturup her bir elemanı 1 artırmak. Size' ı biz giriyoruz
import random

# Kullanıcıdan liste boyutunu al
size = int(input("Liste boyutunu girin: "))

# Random integer liste oluştur (0-100 arası)
arr = [random.randint(0, 100) for _ in range(size)]

print("Orijinal liste:", arr)

# Her elemanı 1 artır
arr = [x + 1 for x in arr]

print("Elemanları 1 artırılmış liste:", arr)
```

```python
# Yukarıdaki işlemin numpy ile yapılmış versiyonu
import numpy as np

# Kullanıcıdan liste boyutunu al
size = int(input("Array boyutunu girin: "))

# Random integer array oluştur (0-100 arası)
arr = np.random.randint(0, 101, size=size, dtype=np.int32)

print("Orijinal array:", arr)

# Her elemanı 1 artır (vektörleştirilmiş)
arr += 1

print("Elemanları 1 artırılmış array:", arr)
```
